```python
ratio = ['speeding','alcohol','not_distracted']
    
for values in ratio:
    df[values].plot(kind='hist')
    plt.title('{valname}'.format(valname = values))
    plt.show()
```

```python
total_case_sort = df.sort_values(by='total',ascending = False)
plt.figure(figsize= (10,10))
sns.barplot(data = total_case_sort,x ='total',y='alcohol')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(data = df,x ='total',y='speeding',kind="bar",height = 4,aspect =3)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(data = df,x ='total',y='ins_losses',kind="bar",height = 4,aspect =3)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.catplot(data = df,x ='total',y='ins_premium',kind="bar",height = 4,aspect =3)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.pairplot(df[['speeding','alcohol','not_distracted']])
```
